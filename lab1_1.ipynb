```python
from PIL import Image
```

```python
img = Image.open("scot_4.jpg")
img
```

```python
img.size
```

```python
img_crop = img.crop((0,0,400,300))
img_crop
```

```python
img_grey = img_crop.convert("L")
img_grey
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage


k = np.array([[0, 0, 0], [0, -1, 0], [0, 0, 0]])
fig, (ax1, ax2) = plt.subplots(1, 2)
result = ndimage.convolve(img_grey, k)
plt.gray()
ax1.imshow(img_crop)
ax2.imshow(result)
plt.show()
```

```python
final = Image.fromarray(result)
final
```

```python
threshold = 100
img_threshold = final.point(lambda x: 255 if x > threshold else 0)
# img_threshold = img_threshold.convert("1")
img_threshold
```

```python
blank = Image.new(color=-1500, size=(400,300), mode="RGB")
img_segmented = Image.composite(img_crop, blank, img_threshold)
img_segmented
```
